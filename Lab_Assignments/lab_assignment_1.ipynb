# NLP Tokenization, Stemming, and Lemmatization (NLTK)

This notebook demonstrates:
- Tokenization: Whitespace, Punctuation-based, Treebank, Tweet, and MWE
- Stemming: Porter Stemmer and Snowball Stemmer
- Lemmatization: WordNet Lemmatizer

```python
import nltk

# Download required resources (safe to run multiple times)
nltk.download('punkt', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)

from nltk.tokenize import (
    wordpunct_tokenize,
    TreebankWordTokenizer,
    TweetTokenizer,
    MWETokenizer,
)
from nltk.stem import PorterStemmer, SnowballStemmer, WordNetLemmatizer
```

```python
text = "Mr. John isn't going to New York today, right? #NLP @OpenAI Let's test tokenization!"
words_for_mwe = ["I", "live", "in", "New", "York", "City", "and", "love", "machine", "learning", "."]

print("Input text:")
print(text)

# 1) Whitespace Tokenization
whitespace_tokens = text.split()
print("\n1) Whitespace Tokens:")
print(whitespace_tokens)

# 2) Punctuation-based Tokenization
punct_tokens = wordpunct_tokenize(text)
print("\n2) Punctuation-based Tokens (wordpunct_tokenize):")
print(punct_tokens)

# 3) Treebank Tokenization
treebank_tokenizer = TreebankWordTokenizer()
treebank_tokens = treebank_tokenizer.tokenize(text)
print("\n3) Treebank Tokens:")
print(treebank_tokens)

# 4) Tweet Tokenization
tweet_tokenizer = TweetTokenizer(preserve_case=False, strip_handles=False, reduce_len=True)
tweet_tokens = tweet_tokenizer.tokenize(text)
print("\n4) Tweet Tokens:")
print(tweet_tokens)

# 5) MWE Tokenization (Multi-Word Expressions)
mwe_tokenizer = MWETokenizer([("New", "York"), ("machine", "learning")], separator="_")
mwe_tokens = mwe_tokenizer.tokenize(words_for_mwe)
print("\n5) MWE Tokens:")
print("Input words:", words_for_mwe)
print("MWE output:", mwe_tokens)
```

```python
sample_words = ["running", "studies", "better", "wolves", "flying", "happiness", "cars"]

porter = PorterStemmer()
snowball = SnowballStemmer("english")
lemmatizer = WordNetLemmatizer()

print("Words:", sample_words)

porter_stems = [porter.stem(w) for w in sample_words]
snowball_stems = [snowball.stem(w) for w in sample_words]
lemmatized_words = [lemmatizer.lemmatize(w) for w in sample_words]

print("\nPorter Stemmer:")
print(porter_stems)

print("\nSnowball Stemmer:")
print(snowball_stems)

print("\nLemmatization (WordNet Lemmatizer):")
print(lemmatized_words)
```
