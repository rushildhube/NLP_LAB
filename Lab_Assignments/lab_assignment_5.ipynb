# Morphology Using Add–Delete Table

## Objective
To study and understand morphology, the structure of words, by analyzing how words are formed from smaller meaning-bearing units called morphemes using an Add–Delete Table approach.

## Introduction / Theory

Morphology is a branch of linguistics that studies how words are formed and structured. A word is made up of morphemes, which are the smallest units of meaning.

### Types of Morphemes
1. Root / Base word - the main meaning of a word, for example, play.
2. Prefix - added at the beginning of a word, for example, un + happy.
3. Suffix - added at the end of a word, for example, play + ing.
4. Inflectional morpheme - shows grammatical change such as tense or number without changing the core meaning.
5. Derivational morpheme - creates a new word or changes the word class.

### Importance of Morphology
Morphology is important in NLP, spell checking, lemmatization, and text analysis.

## Add–Delete Table Concept

The Add–Delete Table explains how new words are formed by:
- Adding prefixes or suffixes
- Deleting letters before adding suffixes
- Changing spelling patterns

This table helps us visualize how a base word changes into a result word through morphological processes.

### Input Base Words
play, happy, write, act, nation, use, create

```python
import pandas as pd

# Base words
words = ["play", "happy", "write", "act", "nation", "use", "create"]

# Add-delete transformation function
def morphology(word):
    if word == "play":
        return (word, "+ing", "-", "playing", "Suffix addition")
    elif word == "happy":
        return (word, "+ness", "y", "happiness", "y -> i rule")
    elif word == "write":
        return (word, "+ing", "e", "writing", "Silent e removal")
    elif word == "act":
        return (word, "+or", "-", "actor", "Derivational suffix")
    elif word == "nation":
        return (word, "+al", "-", "national", "Derivational suffix")
    elif word == "use":
        return (word, "+ful", "-", "useful", "Suffix addition")
    elif word == "create":
        return (word, "+ion", "e", "creation", "e removal")
    else:
        return (word, "-", "-", word, "No change")

# Create table
data = [morphology(word) for word in words]

df = pd.DataFrame(data, columns=["Base Word", "Add", "Delete", "Result Word", "Morphological Process"])
df
```
