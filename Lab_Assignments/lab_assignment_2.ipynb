# Bag-of-Words, TF-IDF, and Word2Vec

This notebook demonstrates:
- Bag-of-Words with count occurrence
- Bag-of-Words with normalized count occurrence
- TF-IDF representation
- Word2Vec embeddings

```python
import re
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from gensim.models import Word2Vec

corpus = [
    "Natural language processing makes text analysis easier",
    "Text analysis and language models are important in NLP",
    "Word embeddings capture semantic relationships between words",
    "Bag of words and TF IDF are common text representation techniques",
]

def preprocess(text):
    text = text.lower()
    text = re.sub(r"[^a-z0-9\s]", "", text)
    return text

processed_corpus = [preprocess(doc) for doc in corpus]
print("Processed corpus:")
for i, doc in enumerate(processed_corpus, start=1):
    print(f"{i}. {doc}")
```

```python
# Bag-of-Words: count occurrence
count_vectorizer = CountVectorizer()
count_matrix = count_vectorizer.fit_transform(processed_corpus)
count_df = pd.DataFrame(
    count_matrix.toarray(),
    columns=count_vectorizer.get_feature_names_out(),
    index=[f"Doc {i}" for i in range(1, len(processed_corpus) + 1)],
)

print("Bag-of-Words (Count Occurrence):")
print(count_df)

# Bag-of-Words: normalized count occurrence
normalized_df = count_df.div(count_df.sum(axis=1), axis=0).fillna(0)
print("\nBag-of-Words (Normalized Count Occurrence):")
print(normalized_df.round(3))

# TF-IDF representation
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(processed_corpus)
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf_vectorizer.get_feature_names_out(),
    index=[f"Doc {i}" for i in range(1, len(processed_corpus) + 1)],
)

print("\nTF-IDF Representation:")
print(tfidf_df.round(3))
```

```python
# Word2Vec embeddings
sentences = [doc.split() for doc in processed_corpus]
word2vec_model = Word2Vec(
    sentences=sentences,
    vector_size=50,
    window=3,
    min_count=1,
    workers=1,
    sg=1,
    epochs=100,
)

sample_words = ["language", "text", "words"]
print("Word2Vec vocabulary:", list(word2vec_model.wv.key_to_index.keys()))

for word in sample_words:
    print(f"\nEmbedding for '{word}':")
    print(np.round(word2vec_model.wv[word], 4))

# Sentence embedding by averaging word vectors
sample_sentence = "language models and text analysis"
sample_tokens = preprocess(sample_sentence).split()
sentence_embedding = np.mean([word2vec_model.wv[token] for token in sample_tokens], axis=0)
print("\nSample sentence:", sample_sentence)
print("Sentence embedding shape:", sentence_embedding.shape)
print(np.round(sentence_embedding[:10], 4))
```
